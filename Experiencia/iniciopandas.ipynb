```python
import pandas as pd
lista_nombres = ["Alexander", "Alejandra", "Dario", "Wilson"]
lista_pesos = [69, 57, 64, 80] # peso en Kg
lista_alturas = [1.78, 1.6, 1.64, 1.71] # altura en m
# convierte el dicc en un dataframe (marco de datos=tabla=dataset=conjunto de datos)
dicc_datos={"Nombre":lista_nombres, "Peso":lista_pesos, "Altura":lista_alturas}
df =pd.DataFrame(dicc_datos)
df


# Importamos la librera pandas, muy utilizada para manejar datos en forma de tablas (DataFrames).
import pandas as pd

# --- PREPARACIN DE LOS DATOS ---

# Nos aseguramos de que cada lista tenga como mximo 100 elementos.
# [:100] significa "tomar desde el inicio hasta el ndice 100 (sin incluirlo)".
# Si la lista tiene ms de 100 valores, la recortamos a 100.
# Si tiene menos de 100, no pasa nada: simplemente toma lo que haya.
lista_nombres = lista_nombres[:100]
lista_pesos = lista_pesos[:100]
lista_alturas = lista_alturas[:100]

# Creamos un diccionario en el que cada clave ser el nombre de una columna
# y cada valor ser la lista de datos que corresponde a esa columna.
dicc_datos = {
    "Nombre": lista_nombres,   # Columna con los nombres de las personas
    "Peso": lista_pesos,       # Columna con los pesos en kilogramos
    "Altura": lista_alturas    # Columna con las alturas en metros
}

# --- CREACIN DEL DATAFRAME ---

# Convertimos el diccionario en un DataFrame de pandas.
# Un DataFrame es como una tabla de Excel: tiene filas (cada persona) y columnas (nombre, peso, altura).
df_2 = pd.DataFrame(dicc_datos)

# --- CLCULO DEL IMC ---

# Calculamos el ndice de Masa Corporal (IMC) para cada persona.
# Frmula: IMC = Peso (kg) / (Altura en metros ^ 2).
# El resultado se almacena en una nueva columna llamada "IMC".
df_2["IMC"] = df_2["Peso"] / (df_2["Altura"] ** 2)

# --- CLASIFICACIN DEL IMC ---

# Definimos una funcin que recibe un valor de IMC y devuelve una categora textual.
# Estas categoras son estndar segn la OMS:
# - < 18.5 = Bajo peso
# - 18.5 - 24.9 = Peso normal
# - 25 - 29.9 = Sobrepeso
# - >= 30 = Obesidad
def clasificar_imc(imc):
    if imc < 18.5:
        return "Peso bajo"
    elif imc < 25:
        return "Normal"
    elif imc < 30:
        return "Sobrepeso"
    else:
        return "Obesidad"

# Aplicamos la funcin "clasificar_imc" a cada valor de la columna "IMC".
# El mtodo .apply() permite ejecutar una funcin sobre cada fila de una columna.
# El resultado se guarda en una nueva columna llamada "Clasificacin".
df_2["Clasificacin"] = df_2["IMC"].apply(clasificar_imc)

# --- VISUALIZACIN DE RESULTADOS ---

# Mostramos las primeras 10 filas de la tabla para verificar que todo sali bien.
# .head(10) devuelve solo las primeras 10 filas (por defecto devuelve 5 si no se le pasa nmero).
print(df_2.head(10))
```

```python
# ==========================================
# 1. Resumen bsico con .describe()
# ==========================================

# El mtodo .describe() en columnas categricas nos da:
# - count: cuntos datos no nulos hay
# - unique: cuntas categoras diferentes existen
# - top: cul es la categora ms frecuente
# - freq: cuntas veces aparece esa categora ms frecuente
print(df_2[["Nombre", "Clasificacin"]].describe())
```

```python
# ==========================================
# 2. Frecuencias absolutas con .value_counts()
# ==========================================

# .value_counts() cuenta cuntas veces aparece cada categora.
# Esto sirve para ver la distribucin de una variable categrica.
# Ejemplo: cuntas personas estn en cada categora de IMC.
print(df_2["Clasificacin"].value_counts())

# Tambin se puede aplicar a "Nombre" (aunque puede haber muchos nombres).
# Aqu mostramos solo los 10 ms frecuentes.
print(df_2["Nombre"].value_counts().head(10))
```

```python
# ==========================================
# 3. Frecuencias relativas (porcentajes)
# ==========================================

# .value_counts(normalize=True) calcula proporciones (entre 0 y 1).
# Multiplicamos por 100 para convertir a porcentajes.
# Esto sirve para entender qu porcentaje de la muestra est en cada categora.
print(df_2["Clasificacin"].value_counts(normalize=True) * 100)
```

```python

# ==========================================
# 4. Tablas cruzadas con pd.crosstab()
# ==========================================

# pd.crosstab() permite cruzar dos variables categricas.
# Aqu, por ejemplo, mostramos cuntas veces aparece cada combinacin
# de Nombre y Clasificacin. (Normalmente til si hay nombres repetidos).
print(pd.crosstab(df_2["Nombre"], df_2["Clasificacin"]))

```

```python

# ==========================================
# 5. Visualizacin de las categoras
# ==========================================

# Con .plot(kind="bar") podemos graficar los conteos de las categoras.
# Esto sirve para visualizar rpidamente la distribucin de la variable.
# En este caso, la distribucin de personas segn su clasificacin de IMC.
df_2["Clasificacin"].value_counts().plot(kind="bar", title="Distribucin de Clasificacin IMC")
```
