```python
import pandas as pd  # Importamos la librería pandas para manejar tablas (DataFrames)

# Nos aseguramos de que las listas tengan exactamente 100 valores cada una
lista_nombres = lista_nombres[:100]   # Cortamos la lista en caso de que tenga más de 100 nombres
lista_pesos = lista_pesos[:100]       # Cortamos la lista en caso de que tenga más de 100 pesos
lista_alturas = lista_alturas[:100]   # Cortamos la lista en caso de que tenga más de 100 alturas

# Creamos un diccionario con las tres listas
dicc_datos = {
    "Nombre": lista_nombres,   # Columna de nombres
    "Peso": lista_pesos,       # Columna de pesos en Kg
    "Altura": lista_alturas    # Columna de alturas en metros
}

# Convertimos el diccionario en un DataFrame (tabla)
df_2 = pd.DataFrame(dicc_datos)

# Calculamos el IMC (Índice de Masa Corporal) y lo guardamos en una nueva columna
# Fórmula: IMC = Peso (kg) / Altura^2 (m^2)
df_2["IMC"] = df_2["Peso"] / (df_2["Altura"] ** 2)

# Definimos una función para clasificar el IMC
def clasificar_imc(imc):
    if imc < 18.5:          # Menos de 18.5 = bajo peso
        return "Peso bajo"
    elif imc < 25:          # Entre 18.5 y 24.9 = normal
        return "Normal"
    elif imc < 30:          # Entre 25 y 29.9 = sobrepeso
        return "Sobrepeso"
    else:                   # 30 o más = obesidad
        return "Obesidad"

# Aplicamos la función a cada fila de la columna IMC
# y creamos una nueva columna "Clasificación"
df_2["Clasificación"] = df_2["IMC"].apply(clasificar_imc)

# Mostramos las primeras 10 filas de la tabla para verificar
print(df_2.head(10))
```

```python
import pandas as pd
lista_nombres = ["Alexander", "Alejandra", "Dario", "Wilson"]
lista_pesos = [69, 57, 64, 80] # peso en Kg
lista_alturas = [1.78, 1.6, 1.64, 1.71] # altura en m
# convierte el dicc en un dataframe (marco de datos=tabla=dataset=conjunto de datos)
dicc_datos={"Nombre":lista_nombres, "Peso":lista_pesos, "Altura":lista_alturas}
df =pd.DataFrame(dicc_datos)
df


# Importamos la librería pandas, muy utilizada para manejar datos en forma de tablas (DataFrames).
import pandas as pd

# --- PREPARACIÓN DE LOS DATOS ---

# Nos aseguramos de que cada lista tenga como máximo 100 elementos.
# [:100] significa "tomar desde el inicio hasta el índice 100 (sin incluirlo)".
# Si la lista tiene más de 100 valores, la recortamos a 100.
# Si tiene menos de 100, no pasa nada: simplemente toma lo que haya.
lista_nombres = lista_nombres[:100]
lista_pesos = lista_pesos[:100]
lista_alturas = lista_alturas[:100]

# Creamos un diccionario en el que cada clave será el nombre de una columna
# y cada valor será la lista de datos que corresponde a esa columna.
dicc_datos = {
    "Nombre": lista_nombres,   # Columna con los nombres de las personas
    "Peso": lista_pesos,       # Columna con los pesos en kilogramos
    "Altura": lista_alturas    # Columna con las alturas en metros
}

# --- CREACIÓN DEL DATAFRAME ---

# Convertimos el diccionario en un DataFrame de pandas.
# Un DataFrame es como una tabla de Excel: tiene filas (cada persona) y columnas (nombre, peso, altura).
df_2 = pd.DataFrame(dicc_datos)

# --- CÁLCULO DEL IMC ---

# Calculamos el Índice de Masa Corporal (IMC) para cada persona.
# Fórmula: IMC = Peso (kg) / (Altura en metros ^ 2).
# El resultado se almacena en una nueva columna llamada "IMC".
df_2["IMC"] = df_2["Peso"] / (df_2["Altura"] ** 2)

# --- CLASIFICACIÓN DEL IMC ---

# Definimos una función que recibe un valor de IMC y devuelve una categoría textual.
# Estas categorías son estándar según la OMS:
# - < 18.5 = Bajo peso
# - 18.5 - 24.9 = Peso normal
# - 25 - 29.9 = Sobrepeso
# - >= 30 = Obesidad
def clasificar_imc(imc):
    if imc < 18.5:
        return "Peso bajo"
    elif imc < 25:
        return "Normal"
    elif imc < 30:
        return "Sobrepeso"
    else:
        return "Obesidad"

# Aplicamos la función "clasificar_imc" a cada valor de la columna "IMC".
# El método .apply() permite ejecutar una función sobre cada fila de una columna.
# El resultado se guarda en una nueva columna llamada "Clasificación".
df_2["Clasificación"] = df_2["IMC"].apply(clasificar_imc)

# --- VISUALIZACIÓN DE RESULTADOS ---

# Mostramos las primeras 10 filas de la tabla para verificar que todo salió bien.
# .head(10) devuelve solo las primeras 10 filas (por defecto devuelve 5 si no se le pasa número).
print(df_2.head(10))
```

```python
# ==========================================
# 1. Resumen básico con .describe()
# ==========================================

# El método .describe() en columnas categóricas nos da:
# - count: cuántos datos no nulos hay
# - unique: cuántas categorías diferentes existen
# - top: cuál es la categoría más frecuente
# - freq: cuántas veces aparece esa categoría más frecuente
print(df_2[["Nombre", "Clasificación"]].describe())
```

```python
# ==========================================
# 2. Frecuencias absolutas con .value_counts()
# ==========================================

# .value_counts() cuenta cuántas veces aparece cada categoría.
# Esto sirve para ver la distribución de una variable categórica.
# Ejemplo: cuántas personas están en cada categoría de IMC.
print(df_2["Clasificación"].value_counts())

# También se puede aplicar a "Nombre" (aunque puede haber muchos nombres).
# Aquí mostramos solo los 10 más frecuentes.
print(df_2["Nombre"].value_counts().head(10))
```

```python
# ==========================================
# 3. Frecuencias relativas (porcentajes)
# ==========================================

# .value_counts(normalize=True) calcula proporciones (entre 0 y 1).
# Multiplicamos por 100 para convertir a porcentajes.
# Esto sirve para entender qué porcentaje de la muestra está en cada categoría.
print(df_2["Clasificación"].value_counts(normalize=True) * 100)
```

```python

# ==========================================
# 4. Tablas cruzadas con pd.crosstab()
# ==========================================

# pd.crosstab() permite cruzar dos variables categóricas.
# Aquí, por ejemplo, mostramos cuántas veces aparece cada combinación
# de Nombre y Clasificación. (Normalmente útil si hay nombres repetidos).
print(pd.crosstab(df_2["Nombre"], df_2["Clasificación"]))

```

```python

# ==========================================
# 5. Visualización de las categorías
# ==========================================

# Con .plot(kind="bar") podemos graficar los conteos de las categorías.
# Esto sirve para visualizar rápidamente la distribución de la variable.
# En este caso, la distribución de personas según su clasificación de IMC.
df_2["Clasificación"].value_counts().plot(kind="bar", title="Distribución de Clasificación IMC")
```
