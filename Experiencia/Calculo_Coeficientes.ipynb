```python
x = [1,2,2,3,4,4,5,6]
y = [2,3,4,4,4,6,5,7]
```

```python
import pandas as pd
df = pd.DataFrame({'x':x,'y':y})
df
```

```python
# Crear la nueva columna
df['x*y'] = df['x'] * df['y']

print(df)
```

```python
# Crear la nueva columna
df['x*2'] = df['x'] ** 2
print(df)
```

```python
# Calcular la suma de todas las columnas
totales = df.sum(numeric_only=True)

# Agregarla como nueva fila
df.loc['Total'] = totales

print(df)
```

```python
m = 8 * (df['x*y']['Total'])
m
```

```python
a1 = 8 * (df['x*y']['Total'])-((df['x']['Total'])*(df['y']['Total']))
print(a1)
b1= 8*(df['x*2']['Total']) - (df['x']['Total'])**2
print(b1)
m=a1/b1
print(m)
```

```python
b=(df['y']['Total']-m*df['x']['Total'])/8
print(b)
```

```python
import matplotlib.pyplot as plt
import numpy as np
y1=m*1+b
print(y1)
plt.plot(y1, 'o') # Plotting y1 as a single point
plt.show() # Display the plot
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Tus datos
x = df['x']
y = df['y']

# Calculamos coeficientes de la recta (pendiente m e intercepto b)
m, b = np.polyfit(x, y, 1)   # ajuste lineal grado 1

# Creamos valores de x para la recta
x_linea = np.linspace(x.min(), x.max(), 100)
y_linea = m * x_linea + b

# Graficamos los puntos y la recta
plt.scatter(x, y, color="blue", label="Datos")
plt.plot(x_linea, y_linea, color="red", label=f"Recta: y = {m:.2f}x + {b:.2f}")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Recta ajustada a tus datos")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# --- Graficar la recta con mis coeficientes m y b ---

import matplotlib.pyplot as plt
import numpy as np

# Valores de x e y del DataFrame
x = df['x']
y = df['y']

# Usar los coeficientes que ya encontraste antes: m y b
# (si en tu notebook ya tienes m y b calculados, aqu no hace falta volver a definirlos)

# Crear valores de x para la recta (desde el mnimo hasta el mximo)
x_linea = np.linspace(x.min(), x.max(), 100)

# Calcular y de la recta con tu m y b
y_linea = m * x_linea + b

# Graficar datos y la recta
plt.scatter(x, y, color="blue", label="Datos")
plt.plot(x_linea, y_linea, color="red", label=f"y = {m:.2f}x + {b:.2f}")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Recta ajustada con mis coeficientes")
plt.legend()
plt.grid(True)
plt.show()
```
