## Cargar los datos

### Subtask:
Cargar el conjunto de datos "Students Social Media Addiction.csv" en un DataFrame de pandas.


**Reasoning**:
Import the pandas library and load the CSV file into a DataFrame named `df`.



```python
import pandas as pd

df = pd.read_csv('Students Social Media Addiction.csv')
display(df.head())
```

## Exploración inicial

### Subtask:
Mostrar las primeras filas del DataFrame, información general sobre las columnas y los tipos de datos, y un resumen estadístico.


**Reasoning**:
Display the first few rows, general information, and descriptive statistics of the DataFrame to understand its structure, data types, and basic statistical properties.



```python
display(df.head())
display(df.info())
display(df.describe())
```

## Limpieza de datos

### Subtask:
Identificar y manejar valores faltantes, duplicados y posibles errores en los datos.


**Reasoning**:
Check for missing values and duplicate rows in the DataFrame.



```python
missing_values = df.isnull().sum()
print("Missing values per column:")
display(missing_values)

duplicate_rows = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicate_rows}")

if duplicate_rows > 0:
    df.drop_duplicates(inplace=True)
    print("Duplicate rows removed.")
```

## Análisis univariado

### Subtask:
Analizar la distribución de cada variable individualmente (histogramas, box plots, etc.).


**Reasoning**:
Generate histograms for the numerical columns to visualize their distributions, then calculate frequencies for categorical columns, and finally generate box plots for numerical columns to identify outliers and dispersion.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Histograms for numerical columns
numerical_cols = ['Age', 'Avg_Daily_Usage_Hours', 'Sleep_Hours_Per_Night', 'Mental_Health_Score', 'Conflicts_Over_Social_Media', 'Addicted_Score']

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 3, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribución de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()

# 2. Frequency for categorical columns
categorical_cols = ['Gender', 'Academic_Level', 'Country', 'Most_Used_Platform', 'Affects_Academic_Performance', 'Relationship_Status']

for col in categorical_cols:
    print(f"\nFrecuencia de categorías para '{col}':")
    display(df[col].value_counts())

# 3. Box plots for numerical columns
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot de {col}')
    plt.xlabel(col)
plt.tight_layout()
plt.show()
```

## Análisis bivariado

### Subtask:
Explorar la relación entre pares de variables (diagramas de dispersión, matrices de correlación, etc.).


**Reasoning**:
Calculate and display the correlation matrix for the numerical columns and create a heatmap visualization of it.



```python
numerical_cols = ['Age', 'Avg_Daily_Usage_Hours', 'Sleep_Hours_Per_Night', 'Mental_Health_Score', 'Conflicts_Over_Social_Media', 'Addicted_Score']
correlation_matrix = df[numerical_cols].corr()
print("Correlation Matrix:")
display(correlation_matrix)

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Matriz de Correlación de Variables Numéricas')
plt.show()
```

**Reasoning**:
Generate scatter plots for key pairs of numerical variables and box plots for relationships between categorical and numerical variables based on the instructions.



```python
# Scatter plots for key numerical variable pairs
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(x=df['Avg_Daily_Usage_Hours'], y=df['Addicted_Score'])
plt.title('Uso Diario vs. Puntuación de Adicción')
plt.xlabel('Horas de Uso Diario Promedio')
plt.ylabel('Puntuación de Adicción')

plt.subplot(1, 2, 2)
sns.scatterplot(x=df['Sleep_Hours_Per_Night'], y=df['Mental_Health_Score'])
plt.title('Horas de Sueño vs. Puntuación de Salud Mental')
plt.xlabel('Horas de Sueño por Noche')
plt.ylabel('Puntuación de Salud Mental')

plt.tight_layout()
plt.show()

# Box plots for categorical and numerical variables
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
sns.boxplot(x=df['Academic_Level'], y=df['Avg_Daily_Usage_Hours'])
plt.title('Nivel Académico vs. Horas de Uso Diario')
plt.xlabel('Nivel Académico')
plt.ylabel('Horas de Uso Diario Promedio')

plt.subplot(2, 2, 2)
sns.boxplot(x=df['Gender'], y=df['Addicted_Score'])
plt.title('Género vs. Puntuación de Adicción')
plt.xlabel('Género')
plt.ylabel('Puntuación de Adicción')

plt.tight_layout()
plt.show()
```

## Visualización de datos

### Subtask:
Crear visualizaciones relevantes para entender mejor los patrones y las relaciones en los datos.


**Reasoning**:
Generate a countplot for the 'Academic_Level' column with specified title and labels, rotating the x-axis labels.



```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Academic_Level', data=df)
plt.title('Distribución por Nivel Académico')
plt.xlabel('Nivel Académico')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()
```
