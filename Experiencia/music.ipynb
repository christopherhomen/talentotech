```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.inspection import permutation_importance

import scipy.stats as stats

# =========================================
# Carga de datos
# =========================================
df = pd.read_csv("/content/songs_normalize.csv")

# =========================================
# Utilidades
# =========================================
def seleccionar_features(df, target="popularity"):
    """
    Selecciona columnas num√©ricas candidatas para predecir la popularidad.
    1) Toma num√©ricas (float/int) excluyendo la variable objetivo.
    2) Filtra columnas con varianza > 0 (evita columnas constantes).
    3) (Opcional) Mantiene columnas t√≠picas de audio si existen.
    """
    num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    if target in num_cols:
        num_cols.remove(target)

    # Quitar columnas con varianza 0
    var_ok = [c for c in num_cols if df[c].var() > 0]

    # Si existen, priorizamos caracter√≠sticas de audio comunes
    candidatas_prioritarias = [
        'danceability', 'energy', 'loudness', 'speechiness', 'acousticness',
        'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms',
        'key', 'mode'
    ]
    features = [c for c in candidatas_prioritarias if c in var_ok]

    # Si por alguna raz√≥n no hay de la lista, usa todas las num√©ricas con varianza>0
    if len(features) == 0:
        features = var_ok

    return features

def print_explicacion_features(features, target="popularity"):
    print("\nüß† ¬øPor qu√© estas columnas?")
    print(f"- Se eligieron variables NUM√âRICAS con varianza > 0, relacionadas con audio,")
    print(f"  candidatas a explicar '{target}'. Ejemplos: danceability, energy, loudness, tempo, etc.")
    print("- Estas variables suelen capturar propiedades medibles de la se√±al musical que")
    print("  influyen en la popularidad (ritmo, energ√≠a, sonoridad, 'felicidad'/valence, etc.).")
    print(f"\nüßæ Variables usadas en el modelo: {features}")

def print_explicacion_errores():
    print("\nüìè ¬øPara qu√© sirven estas m√©tricas en tu data?")
    print("- MSE (Error cuadr√°tico medio): penaliza m√°s los errores grandes; √∫til para comparar modelos.")
    print("- RMSE (Ra√≠z del MSE): mismo significado que MSE pero en unidades de 'popularidad'.")
    print("- MAE (Error absoluto medio): promedio del error absoluto; robusto a outliers.")
    print("- R¬≤ (Coef. de determinaci√≥n): proporci√≥n de variabilidad explicada por el modelo (0 a 1).")

# =========================================
# Entrenamiento base (para reutilizar entre opciones 8, 9 y 10)
# =========================================
def entrenar_modelo_regresion(df, target="popularity", test_size=0.2, random_state=42):
    features = seleccionar_features(df, target)
    print_explicacion_features(features, target)

    X = df[features].copy()
    y = df[target].copy()

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state
    )

    model = LinearRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # M√©tricas
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    print_explicacion_errores()
    print(f"\nüîπ MSE:  {mse:.3f}")
    print(f"üîπ RMSE: {rmse:.3f}")
    print(f"üîπ MAE:  {mae:.3f}")
    print(f"üîπ R¬≤:   {r2:.3f}")

    return model, X_train, X_test, y_train, y_test, y_pred, features

def plot_pred_vs_real(y_test, y_pred):
    plt.figure(figsize=(7,5))
    plt.scatter(y_test, y_pred, alpha=0.5)
    lims = [min(y_test.min(), y_pred.min()), max(y_test.max(), y_pred.max())]
    plt.plot(lims, lims, 'r--', linewidth=1)
    plt.xlabel("Popularidad Real")
    plt.ylabel("Popularidad Predicha")
    plt.title("Predicci√≥n vs Realidad (Popularidad)")
    plt.show()

def plot_hist_residuos(y_test, y_pred):
    residuos = y_test - y_pred
    sns.histplot(residuos, bins=30, kde=True)
    plt.title("Histograma de errores (residuos)")
    plt.xlabel("Error (y_real - y_predicho)")
    plt.show()
    return residuos

def plot_residuos_qq(residuos):
    plt.figure(figsize=(6,6))
    stats.probplot(residuos, dist="norm", plot=plt)
    plt.title("Q‚ÄìQ plot de residuos (¬øgausianos?)")
    plt.show()

def plot_coef_mapping(model, features):
    coefs = pd.Series(model.coef_, index=features).sort_values()
    plt.figure(figsize=(8,5))
    coefs.plot(kind="barh")
    plt.title("Mapping de coeficientes (magnitud y signo)")
    plt.xlabel("Coeficiente lineal")
    plt.ylabel("Feature")
    plt.tight_layout()
    plt.show()

def plot_perm_importance(model, X_test, y_test):
    perm = permutation_importance(model, X_test, y_test, n_repeats=20, random_state=42)
    imp = pd.Series(perm.importances_mean, index=X_test.columns).sort_values()
    plt.figure(figsize=(8,5))
    imp.plot(kind="barh")
    plt.title("Importancia por permutaci√≥n (‚Üì mayor impacto en el error)")
    plt.xlabel("Œî error medio al permutar")
    plt.ylabel("Feature")
    plt.tight_layout()
    plt.show()

# =========================================
# Men√∫ existente + nuevas opciones
# =========================================
def menu():
    print("\n" + "‚ïê" * 46)
    print("      üéµ MEN√ö DE AN√ÅLISIS MUSICAL üéµ")
    print("‚ïê" * 46)
    print(" 1Ô∏è‚É£  Informaci√≥n general del dataset")
    print(" 2Ô∏è‚É£  Top 10 artistas con m√°s canciones")
    print(" 3Ô∏è‚É£  Distribuci√≥n de popularidad")
    print(" 4Ô∏è‚É£  Evoluci√≥n de canciones por a√±o")
    print(" 5Ô∏è‚É£  Correlaci√≥n entre caracter√≠sticas de audio")
    print(" 6Ô∏è‚É£  G√©neros m√°s frecuentes")
    print(" 7Ô∏è‚É£  Canciones m√°s populares")
    print(" 8Ô∏è‚É£  üìà Regresi√≥n lineal (popularidad) + explicaci√≥n y m√©tricas")
    print(" 9Ô∏è‚É£  üó∫Ô∏è Mapping de variables (coeficientes & permutaci√≥n)")
    print(" üîü  üß™ Diagn√≥stico de errores (histograma + Q‚ÄìQ plot)")
    print(" 0Ô∏è‚É£  Salir")
    print("‚ïê" * 46)

def ejecutar(opcion):
    if opcion == 1:
        print("\nüìã Informaci√≥n del DataFrame:")
        print(df.info())
        print("\nüîç Primeras filas del dataset:")
        display(df.head())
        print("\nüìä Estad√≠sticas descriptivas (num√©ricas):")
        display(df.describe())
        print("\nüî† Estad√≠sticas descriptivas (categ√≥ricas):")
        display(df.describe(include=['object', 'bool']))

    elif opcion == 2:
        top_artistas = df['artist'].value_counts().head(10)
        top_artistas.plot(kind="barh", figsize=(8,5))
        plt.title("Top 10 artistas con m√°s canciones")
        plt.show()

    elif opcion == 3:
        sns.histplot(df['popularity'], bins=20, kde=True)
        plt.title("Distribuci√≥n de popularidad")
        plt.show()

    elif opcion == 4:
        canciones_por_a√±o = df['year'].value_counts().sort_index()
        canciones_por_a√±o.plot(figsize=(10,5))
        plt.title("Evoluci√≥n de canciones por a√±o")
        plt.show()

    elif opcion == 5:
        plt.figure(figsize=(10,6))
        sns.heatmap(df.corr(numeric_only=True), annot=False, cmap="coolwarm")
        plt.title("Matriz de correlaci√≥n entre caracter√≠sticas de audio")
        plt.show()

    elif opcion == 6:
        top_genres = df['genre'].value_counts().head(10)
        top_genres.plot(kind="bar", figsize=(8,5))
        plt.title("Top 10 g√©neros m√°s frecuentes")
        plt.show()

    elif opcion == 7:
        top_canciones = df[['artist','song','popularity']].sort_values(by="popularity", ascending=False).head(10)
        print("\nüéµ Top 10 canciones m√°s populares:")
        display(top_canciones)

    elif opcion == 8:
        print("\nüìà Regresi√≥n Lineal: Popularidad en funci√≥n de caracter√≠sticas de audio")
        model, X_train, X_test, y_train, y_test, y_pred, features = entrenar_modelo_regresion(df)
        # Gr√°ficos clave
        plot_pred_vs_real(y_test, y_pred)
        residuos = y_test - y_pred
        sns.histplot(residuos, bins=30, kde=True)
        plt.title("Histograma de errores (residuos)")
        plt.xlabel("Error (y_real - y_predicho)")
        plt.show()

    elif opcion == 9:
        print("\nüó∫Ô∏è Mapping de variables")
        model, X_train, X_test, y_train, y_test, y_pred, features = entrenar_modelo_regresion(df)
        print("\n‚û°Ô∏è  Vista 1: Coeficientes del modelo (magnitud y signo)")
        plot_coef_mapping(model, features)
        print("\n‚û°Ô∏è  Vista 2: Importancia por permutaci√≥n (impacto en el error)")
        plot_perm_importance(model, X_test, y_test)

    elif opcion == 10:
        print("\nüß™ Diagn√≥stico de errores")
        model, X_train, X_test, y_train, y_test, y_pred, features = entrenar_modelo_regresion(df)
        residuos = plot_hist_residuos(y_test, y_pred)
        plot_residuos_qq(residuos)

    elif opcion == 0:
        print("üëã ¬°Gracias por usar el men√∫!")
        return False
    else:
        print("‚ùå Opci√≥n no v√°lida")
    return True

# üîπ Bucle interactivo
while True:
    menu()
    try:
        opcion = int(input("\nüëâ Escribe el n√∫mero de la opci√≥n que deseas: "))
        if not ejecutar(opcion):
            break
    except ValueError:
        print("‚ö†Ô∏è Por favor escribe un n√∫mero v√°lido.")
```
